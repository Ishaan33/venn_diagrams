# How to Create and Customize Venn Diagrams in Python

```python
#Import libraries
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
%matplotlib inline
```

# Venn diagram with two groups

```python
#weighted Venn diagram with two groups
#subsets is a 3-element list (Ab, aB, AB)
venn2(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'));
```

```python
#change colors with the set_colors parameter and vary transparency with alpha parameter
venn2(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'), set_colors=('purple', 'skyblue'), alpha = 0.7);
```

```python
#unweighted Venn diagram with two groups
venn2_unweighted(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'), set_colors=('r', 'g'), alpha = 0.5);
```

## Customizations

```python
#Venn diagram with outline
venn2(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'), set_colors=('r', 'g'), alpha = 0.5);
venn2_circles(subsets = (30, 10, 5));
```

```python
#venn diagram with dashed outline
venn2(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'), set_colors=('r', 'g'), alpha = 0.5);
venn2_circles(subsets = (30, 10, 5), linestyle='dashed', linewidth=2, color='k');
```

```python
#insert title
venn2(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'), set_colors=('r', 'g'), alpha = 0.5);
venn2_circles(subsets = (30, 10, 5), linestyle='dashed', linewidth=2, color='k');
plt.title("I made a Venn Diagram!");
```

# Venn diagram with three groups

```python
venn3(subsets = (20, 10, 12, 10, 9, 4, 3), set_labels = ('Group A', 'Group B', 'Group C'), alpha = 0.5);
```

```python
#To further customize and change the color for a specific area of your overlapping circles, you can use the get_patch_by_id method 
v = venn3(subsets = (20, 10, 12, 10, 9, 4, 3), set_labels = ('Group A', 'Group B', 'Group C'), alpha = 0.5);
v.get_patch_by_id('111').set_color('orange')
#See https://en.wikipedia.org/wiki/Venn_diagram photo if you need help with patch id
```

```python
c=venn3_circles(subsets = (20, 10, 12, 10, 9, 4, 3), linestyle='dashed', linewidth=2, color="grey")
venn3(subsets = (20, 10, 12, 10, 9, 4, 3), set_labels = ('Group A', 'Group B', 'Group C'), alpha = 0.5);
```

```python
c = venn3_circles(subsets = (20, 10, 12, 10, 9, 4, 3), linestyle='dashed', linewidth=1, color="grey")
c[0].set_lw(5.0)
c[1].set_lw(8.0)
c[2].set_lw(2.0)
venn3(subsets = (20, 10, 12, 10, 9, 4, 3), set_labels = ('Group A', 'Group B', 'Group C'), alpha = 0.5);
```
